### Paso 1: Crear un límite para eliminar los datos nulos, mostrar las variables que harán eliminación a sus nulos y mostrar los conteos antes, eliminar los datos nulos y mostrar los conteos de nuevo.

```python
#checar si hay datos nulos, cuantos y ordenarlos.
pd.set_option('display.max_rows', None)  #configurar pandas para ver todas las filas
print(df.isna().sum().sort_values(ascending=False))
```

```python
#Hay que tratar los datos nulos
#Creación del limite y impresion del mismo
limite=len(df)*0.05
print(limite)
```

```python
#Eliminamos valores nulos realizamos una copia del dataSet
columnas=df.columns[df.isna().sum() <= limite]
df_limpio = df.dropna(subset=columnas).copy()

#Verificamos de nuevo los datos nulos por columna
pd.set_option('display.max_rows', 0)  #configurar pandas para ver las filas
print(df_limpio.isna().sum().sort_values(ascending=False))
```

Analizando el resultado despues de la imputacion de filas, podemos llegar a la conclusion que existen columnas con muchos datos nulos las cuales son inutiles para este estudio debido a la cantidad de nulos, estan son las columnas:
#PoolQC           1405
 #MiscFeature      1361
 #Alley            1323
 #Fence            1135

```python
# Eliminación de columnas nulas:

#columna PoolQC
df_limpio = df_limpio.drop("PoolQC", axis=1)

#columna MiscFeature
df_limpio = df_limpio.drop("MiscFeature", axis=1)

#columna Alley
df_limpio = df_limpio.drop("Alley", axis=1)

#columna Fence
df_limpio = df_limpio.drop("Fence", axis=1)

#Verificamos de nuevo los datos nulos por columna
print(df_limpio.isna().sum().sort_values(ascending=False))
```

### Paso 2: Si existen nulos aún, dependiendo de la gráficas anteriores, determinar si hay que imputar por medio de la moda, la mediana o la media; realizar la imputación.

```python
#analizamos las columnas que siguen con nulos y analizamos sus posibles valores.
print("\n FireplaceQu \n ", set(df_limpio['FireplaceQu']))

#analizamos las columnas que siguen con nulos y analizamos sus posibles valores.
print("\n LotFrontage \n ", set(df_limpio['LotFrontage']))

#analizamos las columnas que siguen con nulos y analizamos sus posibles valores.
print("\n GarageFinish \n ", set(df_limpio['GarageFinish']))

#analizamos las columnas que siguen con nulos y analizamos sus posibles valores.
print("\n GarageYrBlt \n ", set(df_limpio['GarageYrBlt']))

#analizamos las columnas que siguen con nulos y analizamos sus posibles valores.
print("\n GarageQual \n ", set(df_limpio['GarageQual']))

#analizamos las columnas que siguen con nulos y analizamos sus posibles valores.
print("\n GarageCond \n ", set(df_limpio['GarageCond']))

#analizamos las columnas que siguen con nulos y analizamos sus posibles valores.
print("\n GarageType \n ", set(df_limpio['GarageType']))
```

Analizando la situacion actual del dataSet se utilizara inputacion por moda y promedio.
en FireplaceQu , GarageFinish , GarageQual , GarageCond y GarageType  se utilizará por moda debido a la similitud de los datos en cada columna y en el caso de la columna LotFrontage y GarageYrBlt la mejor opcion es por promedio ya que son numericas.

```python
#imputación por moda en la columna FireplaceQu 
# Calcular la moda de la columna
moda = df_limpio['FireplaceQu'].mode()[0]
# Imputar los valores nulos con la moda
df_limpio['FireplaceQu'].fillna(moda, inplace=True)

#imputación por moda las columnas restantes para inputación por moda

#columna GarageFinish
moda = df_limpio['GarageFinish'].mode()[0]
df_limpio['GarageFinish'].fillna(moda, inplace=True)

#columna GarageQual
moda = df_limpio['GarageQual'].mode()[0]
df_limpio['GarageQual'].fillna(moda, inplace=True)

#columna GarageCond
moda = df_limpio['GarageCond'].mode()[0]
df_limpio['GarageCond'].fillna(moda, inplace=True)

#columna GarageType
moda = df_limpio['GarageType'].mode()[0]
df_limpio['GarageType'].fillna(moda, inplace=True)

#Verificamos de nuevo los datos nulos por columna
print(df_limpio.isna().sum().sort_values(ascending=False))
```

```python
#Ahora hacemos la inputación por promedio para las dos columnas restantes

# Calcular el promedio de la columna
promedio = df_limpio['LotFrontage'].mean()
# Imputar los valores nulos con el promedio
df_limpio['LotFrontage'].fillna(promedio, inplace=True)

#columna GarageYrBlt
promedio = df_limpio['GarageYrBlt'].mean()
df_limpio['GarageYrBlt'].fillna(promedio, inplace=True)

#Verificamos de nuevo los datos nulos por columna
print(df_limpio.isna().sum().sort_values(ascending=False))
```

### Paso 3: Analizar los datos numéricos, determinar si hay anomalías y utilizar el rango intercuartílico para tratarlos, mostrar gráficos antes y después del tratamiento, debe verse si la distribución se vio afectada.

```python
#Graficamos de nuevo los boxplot para detectar anomalias en los datos numericos
df_numerico = df_limpio.select_dtypes(include=['number'])

for columna in df_numerico.columns:
    plt.figure(figsize=(6, 4))  # Ajustar el tamaño de la figura si es necesario
    sns.boxplot(data=df_numerico[columna], orient="h")  # Crear el boxplot horizontal
    plt.title(f'Boxplot de {columna}')  # Título del gráfico
    plt.show()  # Mostrar el gráfico
```

```python
pd.set_option('display.max_rows', None)  #configurar pandas para ver las filas
display(df_limpio.corr(numeric_only=True))
```

Analizando las correlaciones, nuestra columna objetivo es "SalePrice" por lo que resolver anomalias de columnas sin tanta correlacion con "SalePrice" por eso me enfocare en las que más correlación tienen o correlacion suficiente para afectar los resultadoas finales.

Las columnas con más correlación con "SalePrice" son: OverallQual, YearBuilt, YearRemodAdd, TotalBsmtSF, 1stFlrSF, GrLivArea, FullBath, TotRmsAbvGrd, GarageCars, GarageArea

```python
#Establecemos las nombre de las columnas a tratar anomalias
columnas_a_tratar = ['OverallQual', 'YearBuilt', 'YearRemodAdd', 'TotalBsmtSF', '1stFlrSF', 
                     'GrLivArea', 'FullBath', 'TotRmsAbvGrd', 'GarageCars', 'GarageArea']

#creamos una copia del dataSet para posteriormente vizualizar el cambhio en las anomalias
df_anomalo=df_limpio.copy()
```

```python
factor_iqr = 1.5

# Itera a través de las columnas y elimina las anomalías basadas en el IQR
for columna in columnas_a_tratar:
    Q1 = df_limpio[columna].quantile(0.25)
    Q3 = df_limpio[columna].quantile(0.75)
    IQR = Q3 - Q1
    umbral_superior = Q3 + factor_iqr * IQR
    umbral_inferior = Q1 - factor_iqr * IQR
    
    # Filtra las filas que no cumplen con los umbrales y crea un nuevo DataFrame
    df_limpio = df_limpio[(df_limpio[columna] >= umbral_inferior) & (df_limpio[columna] <= umbral_superior)]
```

```python
#Graficamos de nuevo los boxplot para detectar anomalias en los datos numericos

for columna in columnas_a_tratar:
    plt.figure(figsize=(6, 4))  # Ajustar el tamaño de la figura si es necesario
    sns.boxplot(data=df_anomalo[columna], orient="h")  # Crear el boxplot horizontal
    plt.title(f'ANTES de {columna}')  # Título del gráfico
    plt.show()  # Mostrar el gráfico
    
    plt.figure(figsize=(6, 4))  # Ajustar el tamaño de la figura si es necesario
    sns.boxplot(data=df_limpio[columna], orient="h")  # Crear el boxplot horizontal
    plt.title(f'DESPUES de {columna}')  # Título del gráfico
    plt.show()  # Mostrar el gráfico
```
